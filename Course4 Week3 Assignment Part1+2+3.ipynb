```python

#definition of the condition for color coding
def color_producer(Borough):
    if Borough == 'East Toronto':
        return 'green'
    elif Borough == 'Central Toronto': 
        return 'orange'
    elif Borough == 'Downtown Toronto': 
        return 'blue'
    elif Borough == 'West Toronto': 
        return 'yellow'
    else:
        return 'red' 
```

```python
print('There are {} unique categories.'.format(len(fs['Venue Category'].unique())))
```

```python
# one hot encoding
fs_onehot = pd.get_dummies(fs[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
fs_onehot['Neighborhood'] = fs['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [fs_onehot.columns[-1]] + list(fs_onehot.columns[:-1])
fs_onehot = fs_onehot[fixed_columns]

fs_onehot.head()
```

```python
fs_grouped = fs_onehot.groupby('Neighborhood').mean().reset_index()
fs_grouped.head()
```

```python
#print each neighbourhood with their top 5 venues
num_top_venues = 5

for hood in fs_grouped['Neighborhood']:
    print("----"+hood+"----")
    temp = fs_grouped[fs_grouped['Neighborhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
import numpy as np
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = fs_grouped['Neighborhood']

for ind in np.arange(fs_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(fs_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

# SECTION 3 - Doing k-means clustering

```python
# set number of clusters
kclusters = 10

fs_grouped_clustering = fs_grouped.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(fs_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python

#rename field
#base1['Neighborhood'] = base1['Neighbourhood']
#base1.head()
```

```python
# add clustering labels
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

fs_merged = fs

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
fs_merged = fs_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighborhood')

fs_merged.head() # check the last columns!
```

```python
fs_merged
```

```python
import matplotlib.cm as cm
import matplotlib.colors as colors
# create map
map_clusters = folium.Map(location=[43.6532, -79.3832], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(fs_merged['Neighborhood Latitude'], fs_merged['Neighborhood Longitude'], fs_merged['Neighborhood'], fs_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
